```
import keras
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np 
```

```
model = VGG16(weights="/content/4a05295f-317f-4348-b792-c6c44658e7f7_vgg16_weights_tf_dim_ordering_tf_kernels.h5")
```

```
img = image.load_img('/content/Red-fox-06.jpg', target_size=(224,224)) #note the input size
img_arr = np.expand_dims(image.img_to_array(img), axis=0)
x = preprocess_input(img_arr)
```

```
preds = model.predict(x)

print('Predicted:', decode_predictions(preds, top=3)[0])
```
